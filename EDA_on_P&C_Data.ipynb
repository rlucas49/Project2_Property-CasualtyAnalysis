## Investigating 9998 and 9999 values

```python
# Figure out how many 9998 or 9999 values are present in each column.

dct_missing_values = {}

for i in pdf_data:
    dct_missing_values[i] = {}
    dct_missing_values[i]['9998'] = pdf_data[i].isin([9998]).sum().item()
    dct_missing_values[i]['9999'] = pdf_data[i].isin([9999]).sum().item()
    dct_missing_values[i]['99999'] = pdf_data[i].isin([99999]).sum().item()
dct_missing_values   
```

```python
# Examine the rows that contain 9999 values
# Turn lines on or off as needed

# pdf_data[pdf_data['PREV_POLY_INFORCE_QTY'] == 9999]
# pdf_data[pdf_data['PREV_POLY_INFORCE_QTY'] == 9998]
pdf_data[pdf_data['NB_WRTN_PREM_AMT'] == 9999]
```

### Conclusion Regarding 9998 or 9999 or 99999 values

There does not appear to be very many 9998 or 9999. 
There are many more 99999 values. Replace with NaN.

```python
# Replace rows containing the 9998 or 9999 values
# It's easiest to replace all values with nan and then drop rows containing a null
import numpy as np

print(f"The shape of pdf_data BEFORE inserting nulls: {pdf_data.shape}")

for i in pdf_data:
    pdf_data[i] = pdf_data[i].replace(9998, np.nan)
    pdf_data[i] = pdf_data[i].replace(9999, np.nan)
    pdf_data[i] = pdf_data[i].replace(99999, np.nan)


# pdf_data = pdf_data.dropna()
print(f"The shape of pdf_data AFTER inserting nulls: {pdf_data.shape}")
```

## Examine the categorical variables

```python
# How many unique AGENCY_ID's?

print(f"The number of unique agencies: {pdf_data['AGENCY_ID'].nunique()}")
```

```python
# Examine the categorical variables

l_cat_vars = ['PROD_ABBR','PROD_LINE','STATE_ABBR','VENDOR_IND','VENDOR']
dct_unique = {}

# Get a count of each unique value for each of the variables
for i in l_cat_vars:
    dct_unique[i] = {}
    dct_unique[i]['Values'] = pdf_data[i].unique()
    for k in dct_unique[i]['Values']:
        dct_unique[i][k] = len(pdf_data[pdf_data[i]==k])
dct_unique
```
